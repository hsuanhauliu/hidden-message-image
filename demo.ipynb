This is a demo to show you how the package can be used.

```python
%matplotlib notebook

import hmi
from PIL import Image
import matplotlib.pyplot as plt
```

First, create a message and choose an image that you want to hide your message on. Optionally, you can specify a name for your output image. The default name is "output.png".

```python
img = 'imgs/test1.jpg'
message = "You can't see me!"
encrypted_img = 'imgs/output.png'
```

Encrypt the message by calling the encrypt function.

```python
hmi.encrypt(message, img, output_file=encrypted_img)
```

We can display the two images to see the difference, which you really can't because that's the whole point.

```python
ori_img = Image.open(img)
enc_img = Image.open(encrypted_img)

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(ori_img)
ax2.imshow(enc_img)
fig.show()
```

We can then use decrypt function to decrypt the image to get the secret message.

```python
decrypted_message_1 = hmi.decrypt(img, encrypted_img)
decrypted_message_2 = hmi.decrypt(encrypted_img, img)
print(decrypted_message_1)
print(decrypted_message_2)
```

Note that the order of the image input doesn't matter.
